La regresión lineal permite aproximar una nube de puntos a una línea recta,  intentando que haya la menor distancia entre los puntos y la línea para tener el mínimo error, mientras que la interpolación es encontrar una función que pase por puntos conocidos previamente y que permita obtener los demás datos de los cuales no se tenía información. Luego de ese análisis se puede concluir que la interpolación es el método más adecuado para resolver la problemática del proyecto, debido a que ofrece mayor precisión. En este tipo de interpolación, tomamos coordenadas para obtener una función que pase por estos puntos, en este caso tenemos 16 datos (X, Y) de los cuales no cogeremos 4 debido a que son los faltantes.
Para seguir trabajándolos podemos agruparlos en grupos de 3 o 4 datos e interpolar por tramos (recordando que la función interpoladora nos dará de orden n-1, con n = número de datos). Del mismo modo, podemos agrupar los 12 términos que usaremos para interpolar, recordando que esto nos dará un polinomio de orden n-1 (11) y es muy factible que se presente el fenómeno de Runge ya que el polinomio es muy grande y los puntos son equidistantes.  
Los puntos que se tomaron son en el intervalo (1, 5) en el eje X.  

 <center> $(1, -14) ; (2, -24) ; (3, -32) ; (5, -18)$ </center>

<center> $ X_0=1 ; X_1= 2; X_2= 3; X_3= 5;$ </center>

Con la solución $L(x) = \sum_{j=0}^{n}Y_jl_j(x)$ , donde $L(x)$ va a ser el     polinomio interpolador y $Y_j$ va a ser el punto asociado a cada uno de los     $l_j$.  

Para obtener estos $l_j(x)$ usamos la fórmula:  

<center>
$\prod_{i=0, ij}^{n}\frac{x-x_i}{x_j-x_i} $  
</center>

Así, hallamos el primer polinomio $l_0$,
  
<center>
$l_0= \frac{X-X_1}{X_0-X_1}*\frac{X-X_2}{X_0-X_2}*\frac{X-X_3}{X_0-X_3}=\frac{X-2}{1-2}*\frac{X-3}{1-3}*\frac{X-5}{1-5}$  
</center>  

Entonces,  
<center>
$l_0= \frac{X-2}{-1}*\frac{X-3}{-2}*\frac{X-5}{-4}=\frac{x^3-10x^2+31x-30}{-8}  $ 
</center>

<center>
$l_1= \frac{x^3-9x^2+23x-15}{3}$     

$l_2= \frac{x^3-8x^2+17x-10}{-4}$    

$l_3= \frac{x^3-6x^2+11x-6}{24}$  

</center>    

La fórmula que usamos es:  
<center>
$L(x) = Y_0l_0(x)+Y_1l_1(x)+Y_2l_2(x)+Y_3l_3(x)$  
</center>  

Haciendo todos los debidos cálculos, obtenemos el polinomio interpolador:
<center>

**1.3. Método de Kriging:**

Consiste en métodos geoestadísticos de interpolación que buscan hallar el mejor estimador lineal, el cual debe tener una varianza mínima y que la diferencia entre su esperanza y la del valor a evaluar sea nula, es decir un estimador insesgado. Conociendo los valores obsevados inicialmente y haciendo uso de un  semivariograma para conocer el comportamiento de las variables en cierto espacio, se proceden a estimar los posibles valores faltantes. 

La fórmula para estimar los valores desconocidos con este método, es una suma ponderada de los datos:
 <center> $Z^*(X_0)=\sum_{i=1}^{N}λ_iZ(X_i)$ </center>
 Donde:

 $Z^* =$ valor a estimar

 $X_0 =$ ubicación del valor a estimar

 $Z(X_i) =$ valor medido en ubicación $i$

 $λ_i =$ peso desconocido para valor medido en ubicación $i$

 $N =$ número de valores medidos

 En el semivariograma, se muestran las distancias entre puntos en el eje horizontal y las semivarianzas en el vertical, las cuales están definidas por la siguiente relación:

 <center> $λ(h)= \frac{1}{2N} \sum_{i=1}^{N}[Z(X)-Z(X+h)]^2$ </center>

 Donde:

$λ(h) =$ semivarianza con desfase h

$Z(h) =$ valor de la variable 

$Z(X+h) =$ valor acumulada de la variable hasta el desfase h

$N =$ número de valores medidos
